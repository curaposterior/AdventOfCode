```python
# day 5 part 1

def get_stacks():
    stack1 = ['T', 'Q', 'V', 'C', 'D', 'S', 'N']
    stack2 = ['V', 'F', 'M']
    stack3 = ['M', 'H', 'N', 'P', 'D', 'W', 'Q', 'F']
    stack4 = ['F', 'T', 'R', 'Q', 'D']
    stack5 = ['B', 'V', 'H', 'Q', 'N', 'M', 'F','R']
    stack6 = ['Q', 'W', 'P', 'N', 'G', 'F', 'C']
    stack7 = ['T', 'C', 'L', 'R', 'F', 'W']
    stack8 = ['S', 'N', 'Z', 'T']
    stack9 = ['N', 'H', 'Q', 'R', 'J', 'D', 'S', 'M']
    stacks = [stack1[::-1], stack2[::-1], stack3[::-1], stack4[::-1], stack5[::-1], stack6[::-1], stack7[::-1], stack8[::-1], stack9[::-1]]
    return stacks


def solution1(filename: str, stacks):
    with open(filename, 'r') as f:
        for line in f.readlines():
            line = line.replace('\n','').split() #1, 3, 5
            n = int(line[1])
            mv_from = int(line[3])-1
            mv_to = int(line[5])-1

            try:
                for i in range(n):
                    elem = stacks[mv_from].pop()
                    stacks[mv_to].append(elem)
            except IndexError:
                continue
    flag = ''
    
    for item in stacks:
        flag += item[-1]
    return flag


def solution2(filename: str, stacks):
    
    with open(filename, 'r') as f:
        for line in f.readlines():
            line = line.replace('\n','').split() #1, 3, 5
            n = int(line[1])
            mv_from = int(line[3])-1
            mv_to = int(line[5])-1

            if (n == 1):
                elem = stacks[mv_from].pop()
                stacks[mv_to].append(elem)
            else:
                pom_stack = []
                
                for i in range(n):
                    el = stacks[mv_from].pop(-1)
                    pom_stack.append(el)
                pom_stack = pom_stack[::-1]
                for elem in pom_stack:
                    stacks[mv_to].append(elem)
                pom_stack.clear()
    flag = ''
    
    for item in stacks:
        flag += item[-1]
    return flag

stacks1 = get_stacks()
stacks2 = get_stacks()
print("Flag 1", solution1('input_5.txt', stacks1))
print("Flag 2", solution2('input_5.txt', stacks2))
```
