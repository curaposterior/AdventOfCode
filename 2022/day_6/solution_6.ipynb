```python
# day 6 part 1


def find_packet_marker(buffer: str, length: int):
    for i in range(len(buffer)):
        elems = set()
        test_buff = buffer[i:]

        for j in range(length):
            if test_buff[j] in elems:
                break
            elems.add(test_buff[j])

        if len(elems) == length:
            return i + length
    
    
def solution_6(filename: str, length: int):
    with open(filename, 'r') as f:
        strings = f.read().split()
        for string in strings:
            print(find_packet_marker(string, length))
solution_6('input_6.txt', 4)
solution_6('input_6.txt', 14)
```
