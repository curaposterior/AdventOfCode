```python
# day 3 part 1

def check_priority(item: str):
    val = ord(item)
    if val >= 97 and val <= 122:
        return 1 + val - ord("a")
    else:
        return 27 + val - ord("A") 

def find_same_characters(string1: str, string2: str):
    string1, string2 = set(string1), set(string2)
    score = 0
    for letter in string2:
        if letter in string1:
            score += check_priority(letter)
    return score



def solution(filename):
    score = 0
    with open(filename, 'r') as f:
        for line in f.readlines():
            line = line.replace('\n', '')
            pivot = int(len(line)/2)
            score += find_same_characters(line[:pivot], line[pivot:])

    return score

# solution('test_input_2.txt')
solution('input_3.txt')
```

```python
# day 3 part 2

def solution_2(filename):
    score = 0
    with open(filename, "r") as f:
        backpacks = [set(line.strip()) for line in f.readlines()]
        for i in range(0, len(backpacks), 3):
            for item in backpacks[i]:
                if (item in backpacks[i+1]) and (item in backpacks[i+2]):
                    score += check_priority(item)
                    break
    return score


# solution_2('test_input_3.txt')
solution_2('input_3.txt')
```
