```python
'''
1 for Rock, 2 for Paper, and 3 for Scissors
0 if you lost, 3 if the round was a draw, and 6 if you won
'''

def part_one(their_throw, your_response):
    round_points = 0
    if their_throw == 'A': # Rock
        if your_response == 'X': # Rock (+1)
            round_points += 4 # Draw (+3)
        if your_response == 'Y': # Paper (+2)
            round_points += 8 # Win (+6)
        if your_response == 'Z': # Scissors (+3)
            round_points += 3 # Loss (+0)
 
    elif their_throw == 'B': # Paper
        if your_response == 'X': # Rock (+1)
            round_points += 1 # Loss (+0)
        if your_response == 'Y': # Paper (+2)
            round_points += 5 # Draw (+3)
        if your_response == 'Z': # Scissors (+3)
            round_points += 9 # Win (+6)    
   
    elif their_throw == 'C': # Scissors
        if your_response == 'X': # Rock (+1)
            round_points += 7 # Win (+6)
        if your_response == 'Y': # Paper (+2)
            round_points += 2 # Loss (+0)
        if your_response == 'Z': # Scissors (+3)
            round_points += 6 # Draw (+3)
    return round_points

def solution(filename):
    score = 0
    with open(filename, 'r') as f:
        for line in f.readlines():
            line = line.split()
            score += part_one(line[0], line[1])
    return score
fname = 'input.txt'
solution(fname)
```

```python
# part two
#X means you need to lose, Y means you need to end the round in a draw, and Z means you need to win. Good luck!"
'''
1 for Rock X, 2 for Paper Y, and 3 for Scissors Z
0 if you lost, 3 if the round was a draw, and 6 if you won
'''

def part_two(their_throw, your_response):
    round_points = 0
    if their_throw == 'A': # Rock
        if your_response == 'X': # LOSE (+1)
            round_points += 3 # Draw (+3)
        if your_response == 'Y': # DRAW (+2)
            round_points += 4 # Win (+6)
        if your_response == 'Z': # WIN (+3)
            round_points += 8 # Loss (+0)
 
    elif their_throw == 'B': # Paper
        if your_response == 'X': #  LOSE (+1)
            round_points += 1 # Loss (+0)
        if your_response == 'Y': #  DRAW (+2)
            round_points += 5 # Draw (+3)
        if your_response == 'Z': # WIN (+3)
            round_points += 9 # Win (+6)    
   
    elif their_throw == 'C': # Scissors
        if your_response == 'X': # LOSE (+1)
            round_points += 2 # Win (+6)
        if your_response == 'Y': # DRAW (+2)
            round_points += 6 # Loss (+0)
        if your_response == 'Z': # WIN (+3)
            round_points += 7 # Draw (+3)
    return round_points

def solution_2(filename):
    score = 0
    with open(filename, 'r') as f:
        for line in f.readlines():
            line = line.split()
            score += part_two(line[0], line[1])
    return score

print(solution_2('test_input.txt'))
print(solution_2('input.txt'))
```
