```python
# day 4 part 1

def check_if_consist(string1: str, string2: str) -> bool:
    #check if first consists second
    string1, string2 = string1.split('-'), string2.split('-')
    
    if (int(string1[0]) <= int(string2[0]) and int(string1[1]) >= int(string2[1])):
        return True

    #check if second consists first 
    if (int(string2[0]) <= int(string1[0]) and int(string2[1]) >= int(string1[1])):
        return True

    return False

def solution(filename: str) -> int:
    counter = 0
    with open(filename, 'r') as f:
        for line in f.readlines():
            line = line.replace('\n','').split(',')
            if (check_if_consist(line[0], line[1])):
                counter += 1
    

    return counter

solution('input_4.txt')
```

```python
# day 4 part 2

def check_if_overlaps(string1: str, string2: str) -> bool:
    nums1, nums2 = string1.split('-'), string2.split('-')

    for i in range(int(nums1[0]), int(nums1[1])+1):
        if i in range(int(nums2[0]), int(nums2[1])+1):
            return True
    
    return False


def solution2(filename: str) -> int:
    counter = 0
    with open(filename, 'r') as f:
        for line in f.readlines():
            line = line.replace('\n','').split(',')
            if (check_if_overlaps(line[0], line[1])):
                counter += 1
    return counter

solution2('input_4.txt')
```
