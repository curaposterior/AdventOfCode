```python
#day 1
def solution_1(filename: str) -> int:
    max_val = 0
    suma = 0
    with open(filename, 'r') as f:
        for i in f.readlines():
            try:
                suma += int(i.replace('\n', ''))
            except ValueError:
                if (suma > max_val):
                    max_val = suma
                suma = 0
    return max_val

solution_1('input_1.txt')
```

```python
#top 3 max, priority queue
from heapq import heappush, heappop

def solution_2(filename: str) -> int:
    q = []
    suma = 0
    with open(filename, 'r') as f:
        for i in f.readlines():
            try:
                suma += int(i.replace('\n', ''))
            except ValueError:
                heappush(q, (-suma))
                suma = 0
    val1 = heappop(q)
    val2 = heappop(q)
    val3 = heappop(q)
    return -(val1 + val2 + val3)
solution_2('input_1.txt')
```
