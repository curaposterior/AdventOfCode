```python
# day 10 part 1

def solution(filename):
    with open(filename, 'r') as f:
        
        X = 1
        cycle = 0
        sum_of_cycles = 0
        for line in f.readlines():
            cycle += 1

            if (cycle == 20 or cycle == 60 or cycle == 100 or cycle == 140 or cycle == 180 or cycle == 220):
                n = cycle * X
                sum_of_cycles += n

            line = line.split() #operation/value
            
            if line[0] == "noop":
                continue
            
            if line[0] == "addx":
                cycle += 1
                if (cycle == 20 or cycle == 60 or cycle == 100 or cycle == 140 or cycle == 180 or cycle == 220):
                    n = cycle * X
                    sum_of_cycles += n
                X = X + int(line[1])
                
        


    print(sum_of_cycles)

            

solution('input_10.txt')
# solution('input_10.txt')
```
